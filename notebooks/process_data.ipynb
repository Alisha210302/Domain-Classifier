```python
import pandas as pd
import re
```

### Job Profile Dataset

```python
jp_df = pd.read_csv('../Dataset/raw data/job_profile.csv')
jp_df1 = pd.read_csv('../Dataset/raw data/job_profile1.csv')
jp_df2 = pd.read_csv('../Dataset/raw data/job_profile2.csv')
jp_df3 = pd.read_csv('../Dataset/raw data/job_profile3.csv')
jp_df4 = pd.read_csv('../Dataset/raw data/job_profile4.csv')
```

```python
def clean_text(text):
    # Remove \n, special characters, and numbers
    text = re.sub(r'\\n', ' ', text)  # Replace \n with a space
    text = re.sub(r'[^\w\s]', '', text)  # Remove special characters
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

jp_df = jp_df[['job_title', 'description_text']]
jp_df['job_title'] = jp_df['job_title'].apply(clean_text)
jp_df['description_text'] = jp_df['description_text'].apply(clean_text)
jp_df.dropna(inplace=True)
jp_df.columns = ['job_title', 'description']
jp_df.head()
```

```python
jp_df1 = jp_df1[['job_title', 'job_overview']]
jp_df1['job_title'] = jp_df1['job_title'].apply(clean_text)
jp_df1['job_overview'] = jp_df1['job_overview'].apply(clean_text)
jp_df1.dropna(inplace=True)
jp_df1.columns = ['job_title', 'description']
jp_df1.head()
```

```python
jp_df2 = jp_df2[['job_title', 'job_summary']]
jp_df2.dropna(inplace=True)
jp_df2.columns = ['job_title', 'description']
jp_df2.head()
```

```python
jp_df3 = jp_df3[['category', 'description']]
jp_df3.dropna(inplace=True)
jp_df3.columns = ['job_title', 'description']
jp_df3.sample(5)
```

```python
jp_df4 = jp_df4[['Job Title', 'Job Description']]
jp_df4.dropna(inplace=True)
jp_df4.columns = ['job_title', 'description']
jp_df4.sample(5)
```

```python
df_jp = pd.concat([jp_df, jp_df1, jp_df2, jp_df3, jp_df4], axis=0, ignore_index=True)
df_jp['domain'] = 'Job Profile'
df_jp.columns = ['sub_domain','description', 'domain']
df_jp = df_jp[['description', 'domain', 'sub_domain']]
df_jp.sample(10)

df_jp.to_csv('../Dataset/processed_data/processed_job_profile.csv')
```

### News Dataset

```python
news_df = pd.read_csv('../Dataset/raw data/news.csv')
news_df1 = pd.read_csv('../Dataset/raw data/news1.csv')
news_df2 = pd.read_csv('../Dataset/raw data/news2.csv')
```

```python
news_df.dropna(inplace=True)
news_df.head()
```

```python
news_df1 = news_df1[['Text', 'Category']]
news_df1.columns = ['data', 'labels']
news_df1.dropna(inplace=True)
news_df1.head()
```

```python
news_df2 = news_df2[['short_description', 'category']]
news_df2.columns = ['data', 'labels']
news_df2['labels'] = news_df2['labels'].str.lower()
news_df2.dropna(inplace=True)
news_df2.head()
```

```python
df_news = pd.concat([news_df, news_df1, news_df2],
                  axis=0, ignore_index=True)
df_news['domain'] = 'News'
df_news.columns = ['description', 'sub_domain', 'domain']
df_news = df_news[['description', 'domain', 'sub_domain']]
df_news.sample(10)

df_news.to_csv('../Dataset/processed_data/processed_news.csv')
```

### Ecommerce Dataset

```python
df_ec = pd.read_csv('../Dataset/raw data/ecommerce.csv')
df_ec['domain'] = 'Ecommerce'
df_ec.columns = ['sub_domain', 'description', 'domain']
df_ec = df_ec[['description', 'domain', 'sub_domain']]
df_ec.dropna(inplace=True)
df_ec.head()

df_ec.to_csv('../Dataset/processed_data/processed_ecommerce.csv')
```

### Medical Dataset

```python
df_med = pd.read_csv('../Dataset/raw data/medical.csv')

df_med.columns = ['domain', 'sub_domain', 'description']
df_med = df_med[['description', 'domain', 'sub_domain']]
df_med.dropna(inplace=True)
df_med.head()

df_med.to_csv('../Dataset/processed_data/processed_medical.csv')
```

### Final Dataset

```python
df_ecom = pd.read_csv('../Dataset/processed_data/processed_ecommerce.csv')
df_ecom = df_ecom.reset_index(drop=True)
df_med = pd.read_csv('../Dataset/processed_data/processed_medical.csv')
df_med = df_med.reset_index(drop=True)
df_news = pd.read_csv('../Dataset/processed_data/processed_news.csv')
df_news = df_news.reset_index(drop=True)

df_final = pd.concat([df_ecom, df_med, df_news],
                     axis=0, ignore_index=True).drop(columns=['Unnamed: 0'])
df_final.sample(10)

df_final.to_csv('../Dataset/processed_data/final_dataset.csv', index=False)
```
